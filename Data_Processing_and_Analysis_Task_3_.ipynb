# ***Problem 1: Employee Performance Bonus Eligibility***

```python
# Create the dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find the highest score
highest_score = max(employees.values())

# Find employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
# Print the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

# ***Problem 2: Search Query Keyword Analysis***

```python
# Store the sentence
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Split sentence into words
words = query.split()

# Count frequency of each word
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

result = {}

# Display only words that appear more than once
for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

# ***Problem 3: Sensor Data Validation***

```python
# Store the sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create an empty list to store valid readings
valid_readings = []

# Loop through the list using index
for index in range(len(sensor_readings)):

    # Check if the reading is even
    if sensor_readings[index] % 2 == 0:
        valid_readings.append((index, sensor_readings[index]))

# Print result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# ***Problem 4: Email Domain Usage Analysis***

```python
# Store emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Create dictionary to store domain count
domain_count = {}

# Extract domains and count them
for email in emails:
    domain = email.split("@")[1]   # Get part after @

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate percentage and display
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

# ***Problem 5: Sales Spike Detection***

```python
# Store daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average = sum(sales) / len(sales)

# Calculate threshold (30% above average)
threshold = average * 1.3

# Detect and display spike days
for index in range(len(sales)):
    if sales[index] > threshold:
        print(f"Day {index + 1}: {sales[index]}")
```

# ***Problem 6: Duplicate User ID Detection***

```python
# Store user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Create dictionary to count occurrences
id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Display duplicates only
print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
