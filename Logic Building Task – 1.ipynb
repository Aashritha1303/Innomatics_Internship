***2. Pass / Fail Analyzer***

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for i in marks:
    if i>=50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

***3. Simple Data Cleaner***

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for i in names:
    cleaned_name = i.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
```

***4. Message Length Analyzer***

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for i in messages:
    print("Message:", i)
    print("Length:", len(i))
    if len(i) > 10:
        print("Message longer than 10 characters")
    print()
```

***5. Error Message Detector***

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for i in logs:
    if i == "ERROR":
        error_count += 1
print("Number of ERROR entries:", error_count)
```
