# ***Problem 1: Smart Light Controller***

```python
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(self.name, "is", self.status)


# Taking input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.show_status()
```

# ***Problem 2: Employee ID Card System***

```python
class Employee:

    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display()
```

# ***Problem 3 – Mobile Contact Record***

```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


name = input("Contact Name: ")
phone = input("Phone Number: ")

c = Contact(name, phone)
c.save_contact()
```

# ***Problem 4 – Product Price Tag Generator***

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def show_price(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


name = input("Product Name: ")
price = int(input("Price: "))

p = Product(name, price)
p.show_price()
```

# ***Problem 5 – Movie Rating Display System***

```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


name = input("Movie Name: ")
rating = float(input("Rating: "))

m = Movie(name, rating)
m.display()
```

# ***Problem 6 – Delivery Address Manager***

```python
class Delivery:

    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


customer = input("Customer Name: ")
address = input("Address: ")

d = Delivery(customer, address)
d.show_details()
```
