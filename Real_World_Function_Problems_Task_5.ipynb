# ***Problem 1: Post Engagement Analyzer***

```python
def check_engagement(likes_list):
    total = 0

    for like in likes_list:
        total = total + like

    print("Total Likes:", total)

    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes = [200, 150, 300, 400]
check_engagement(likes)
```

# ***Problem 3: Rainfall Checker***

```python
def rainfall_check(rain_list, required):
    total = 0

    for r in rain_list:
        total = total + r

    avg = total / len(rain_list)

    print("Average Rainfall:", avg)

    if avg >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rain_data = [60, 80, 70, 78]
rainfall_check(rain_data, 70)
```

# ***Problem 4: Duplicate Account Detection***

```python
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["ram", "sita", "ram", "arjun"]
check_duplicates(users)
```

# ***Problem 6: Premium Crop Filter***

```python
def premium_filter(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
premium_filter(crop_prices)
```

# ***Problem 7: Application Health Checker***

```python
def health_check(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


health_check(7)
```

# ***Problem 9: Attendance Eligibility***

```python
def attendance_check(att_list):
    present = 0

    for day in att_list:
        if day == "P":
            present = present + 1

    percent = (present / len(att_list)) * 100

    print("Attendance Percentage:", percent)

    if percent >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
attendance_check(attendance)
```

# ***Problem 10: Smart Electricity Bill Analyzer***

```python
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


total_bill, usage = calculate_bill(250)

print("Total Bill:", total_bill)
print("Usage Level:", usage)
```
