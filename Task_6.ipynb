# ***Problem Statement 1: Apartment Visitor Tracking System***

```python
def visitor_tracker(limit, entry_logs):
    count = 0
    max_count = 0

    for item in entry_logs:
        if item == "ENTER":
            count += 1
        elif item == "EXIT":
            if count > 0:
                count -= 1

        if count > max_count:
            max_count = count

    print("Visitors Currently Inside:", count)

    if count > limit:
        print("Status: Limit Crossed")
    else:
        print("Status: Under Control")

    print("Maximum Visitors Today:", max_count)


limit = 40
logs = ["ENTER", "ENTER", "EXIT", "ENTER", "ENTER", "EXIT"]

visitor_tracker(limit, logs)
```

# ***Problem Statement 2: Grocery Delivery Time Calculator***

```python
def delivery_time(distance, traffic_type, rain_status):
    total_time = distance * 4

    if traffic_type == "Low":
        total_time += 5
    elif traffic_type == "High":
        total_time += 15

    if rain_status == "Yes":
        total_time += 10

    return total_time


distance = 6
traffic = "High"
rain = "Yes"

result = delivery_time(distance, traffic, rain)
print("Total Delivery Time:", result, "minutes")
```

# ***Problem Statement 3: Stadium Seat Booking Checker***

```python
def check_occupancy(total_seats, booked_number):
    percentage = (booked_number / total_seats) * 100
    print("Booking Percentage:", int(percentage), "%")

    if percentage == 100:
        print("Stadium Status: Fully Booked")
    elif percentage >= 60:
        print("Stadium Status: Nearly Full")
    else:
        print("Stadium Status: Seats Available")


total = 150
booked = 100

check_occupancy(total, booked)
```

# ***Problem Statement 4: Computer CPU Usage Monitor***

```python
def cpu_monitor(readings):
    total = 0

    for r in readings:
        total += r

    avg = total / len(readings)

    print("Average CPU Usage:", int(avg), "%")

    if avg < 50:
        print("Condition: Safe")
    elif avg <= 75:
        print("Condition: Moderate")
    else:
        print("Condition: High Load")


values = [40, 55, 60, 70]

cpu_monitor(values)
```

# ***Problem Statement 5: School Electricity Usage Checker***

```python
def electricity_check(data):
    high_usage = []

    for item, hrs in data.items():
        if hrs > 6:
            high_usage.append(item)

    if len(high_usage) > 0:
        print("High Usage Items:", ", ".join(high_usage))
        print("Alert: Required")
    else:
        print("All usage normal")
        print("Alert: Not Required")


usage_data = {
    "Fan": 4,
    "AC": 8,
    "Lights": 5
}

electricity_check(usage_data)
```

# ***Problem Statement 6: Seminar Registration System***

```python
def seminar_registration(max_limit, applied_students):
    confirmed = 0
    waiting = 0

    for i in range(applied_students):
        if confirmed < max_limit:
            confirmed += 1
        else:
            waiting += 1

    print("Confirmed Students:", confirmed)
    print("Waiting List:", waiting)

    if confirmed == max_limit:
        print("Registration Closed")
    else:
        print("Registration Open")


limit = 80
applied = 90

seminar_registration(limit, applied)
```
