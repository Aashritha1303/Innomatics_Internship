# ***Problem 1: Mobile Recharge Validation System***

```python
def check_recharge(amount):
    plans = [199, 299, 399, 599]

    if amount >= 50 and amount in plans:
        return True
    else:
        return False


while True:
    amt = int(input("Enter recharge amount: "))

    if check_recharge(amt):
        print("Recharge done successfully")
        break
    else:
        print("Invalid amount or plan. Please try again.")
```

# ***Problem 2: Inventory Reorder Alert System***

```python
def inventory_check(items):
    for name, qty in items.items():
        if qty < 15:
            print(name, "- Reorder Alert")
        else:
            print(name, "- Stock OK")


products = {
    "Pen": 20,
    "Notebook": 10,
    "Bag": 5,
    "Bottle": 25
}

inventory_check(products)
```

# ***Problem 3: Student Result Processing System***

```python
def result_check(marks):
    total = 0

    for m in marks:
        total = total + m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 45, 55, 70]
print("Result:", result_check(marks_list))
```

# ***Problem 4: Cab Fare Estimator***

```python
def fare_calculate(km, peak):
    fare = 50 + (12 * km)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance: "))
    peak_time = input("Peak hour? (yes/no): ").lower()

    if distance > 0:
        print("Total Fare:", fare_calculate(distance, peak_time))
        break
    else:
        print("Distance should be greater than 0. Try again.")
```

# ***Problem 5: Attendance Eligibility Checker***

```python
def attendance_check(att_list):
    present = 0

    for day in att_list:
        if day == "P":
            present = present + 1

    percent = (present / len(att_list)) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "P", "A", "P"]
print("Status:", attendance_check(attendance))
```

# ***Problem 6: Password Strength Checker***

```python
def password_check(pwd):
    digit = False
    special = False
    special_chars = "@#$"

    if len(pwd) < 8:
        return "Weak Password"

    for ch in pwd:
        if ch.isdigit():
            digit = True
        if ch in special_chars:
            special = True

    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(password_check(password))
```
