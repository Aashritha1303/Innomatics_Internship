***Problem 1: Unique Words in a Sentence***

```python
# Take input from the user
sentence = input("Enter sentence: ")

# Convert the sentence into a list of words
# split() breaks the sentence wherever there is a space
words = sentence.split()

# Convert the list into a set
# Set automatically removes duplicate words
unique_words = set(words)

# Count the number of unique words
count = len(unique_words)

# Print the results
print("Unique words count:", count)
print("Unique words:", unique_words)
```

***Problem 2: Highest Salary from Employee Data***

```python
# Store employee data in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee with highest salary
# max() checks values when we use key=employees.get
highest_employee = max(employees, key=employees.get)

# Get the highest salary using the name
highest_salary = employees[highest_employee]

# Print the result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

***Problem 3: Find Maximum and Minimum Values***

```python
# Store the list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    # Check for maximum
    if num > maximum:
        maximum = num

    # Check for minimum
    if num < minimum:
        minimum = num

# Print the results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

***Problem 4: Count Products Above a Price Threshold***

```python
# Store the list of prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through each price
for price in prices:
    # Check if price is greater than 1000
    if price > 1000:
        count += 1  # Increase count if condition is true

# Print result
print("Products above 1000:", count)
```

***Problem 5: Calculate Attendance Percentage***

```python
# Store attendance data
attendance = ["P", "P", "A", "P", "P"]

# Count number of "P"
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
total_days = len(attendance)
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

***Problem 7: Count Character Frequency***

```python
# Store the string
text = "python"

# Create empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    # If character already exists in dictionary
    if ch in char_count:
        char_count[ch] += 1
    else:
        # If character appears first time
        char_count[ch] = 1

# Print result
print(char_count)
```

***Problem 9: Check if a Key Exists in a Dictionary***

```python
# Create a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Define the key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

***Problem 10: Calculate Average Marks***

```python
# Store the marks
marks = [70, 75, 80, 65]   # Example marks

# Calculate total marks
total = 0

for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

# Print result
print("Average Marks:", average)
```
